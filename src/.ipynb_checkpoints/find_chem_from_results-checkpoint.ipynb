```python
infile = "../data/NER_data/BC5CDR-chem/test.tsv"

list_words = []

with open(infile) as f:
    lines = f.readlines()

chem = ""
for line in lines:
    line = line.split()
    
    if len(line)!= 0:
        if line[-1]=="O":
            if len(chem) != 0:
                list_words.append(chem)
                chem=""
            continue
        elif line[-1] == "B":
            chem+="".join(line[:-1])
        elif line[-1] == "I":
            chem+="".join(line[:-1])
    #         chem+=line[:-1]
    
# print(list_words)
len(list_words), len(set(list_words))
```
